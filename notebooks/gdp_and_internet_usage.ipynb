Question 7: 202 Countries have data for all years (6666 entries in cell 22, / 33 for number of years in table). 'Cayman Islands', 'Djibouti', 'Kosovo',
       'Sint Maarten (Dutch part)', 'Somalia', 'Turks and Caicos Islands'are the countries missing many entries (Searched for countries missing 15 or more).
```python
country_count = gdp_df.Country.value_counts()
country_count = country_count.to_frame()
country_count = country_count.reset_index()
```

```python
gdp_df = gdp_df.sort_values(by = 'country_count')
gdp_df
```

Question 21: The three countries with the highest internet use percent in 2014 were iceland, falkland islands, and bermuda. The facet grid shows that Bermuda and Iceland have similar gdp trends, with iceland seeming more stable. The Falkland Islands are missing GDP data for every year. Looking back at question 7, it seems like the falkland islands are not included in the gdp dataframe. 

```python
top_3_2014 = gdp_and_internet_use.loc[gdp_and_internet_use['Year'] == 2014].nlargest(3, 'Internet_Users_Pct')
```

```python
top_3_2014 = top_3_2014.drop(columns = ['Year', 'Internet_Users_Pct', 'GDP_Per_Capita', 'country_count']).reset_index(drop = True)
```

```python
top_3_2014_all = pd.merge(top_3_2014, gdp_and_internet_use,
                          left_on = 'Country', right_on = 'Country',
                          how = 'left')
```

```python
falkland = top_3_2014_all.loc[top_3_2014_all['Country'] == 'Falkland (Malvinas) Islands']
falkland
```

```python
facetgrid = sns.FacetGrid(top_3_2014_all, col = "Country")
facetgrid.map_dataframe(sns.scatterplot, x = 'Year', y = 'GDP_Per_Capita')
```
